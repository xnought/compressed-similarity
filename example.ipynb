```python
import pandas as pd
from gzip_similarity import gzip_similarity
import numpy as np
```

```python
df = pd.read_parquet("./data/prompts_10k.parquet")
```

```python
prompts = df["prompt"].tolist()
responses = df["response"].tolist()
```

```python
sims = np.zeros((1, len(prompts)))
```

```python
sims
```

```python
def mapped_similarity(a:str, bs: list[str]):
    sims = np.zeros((1, len(bs)))
    for i in range(len(bs)):
        sims[0][i] = gzip_similarity(a, bs[i])
    return sims
```

```python
output = mapped_similarity("A funny battle", prompts)
sorted = np.argsort(output)
sorted[:10]
df.iloc[sorted[:10][0],:]["prompt"].tolist()
```
